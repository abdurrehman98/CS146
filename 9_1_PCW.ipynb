```
import scipy.stats as sts
import numpy as np
import random
import matplotlib.pyplot as plt
```

```
alpha = 1
beta = 1
#the range I will set is 35 since high scoring nba players tend to average 25 - 30 ppg (points per game)
n = [[random.randint(1, 35) for i in range(4)] for j in range(3)]
# using n to model
p = sts.beta.rvs(alpha, beta, size=3)
k = [[sts.binom.rvs(n[i][j],p[i]) for j in range(4)] for i in range(3)]
```
